**Part A**

The gain of a transimpedance amplifier with finite bandwidth op amp is:

$$G(s) = \frac{v_{out}}{i_{in}} = Z_{f} \cdot \frac{A(s)}{1 + A(s) + s C_{in} Z_{f}}$$

with $A(s) = \frac{A_{0}}{1 + \frac{s}{s_{p}}} = \frac{\omega _{u}}{s + s_{p}}$ and $Z_{f} = R_{f} \parallel \frac{1}{s C_{f}} = \frac{R_{f}}{1 + s R_{f} C_{f}}$.

$$G(s) = \frac{R_{f}}{1 + s R_{f} C_{f}} \cdot \frac{\frac{A_0}{1 + \frac{s}{s_p}}}{1 + \frac{A_0}{1+\frac{s}{s_p}} + s C_{in} \frac{R_{f}}{1 + s R_{f} C_{f}}}$$

$$ = \frac{R_{f}}{1 + s R_{f} C_{f}} \cdot \frac{A_0}{1 + \frac{s}{s_p} + A_0 + \left( 1 + \frac{s}{s_p} \right) s C_{in} \frac{R_{f}}{1 + s R_{f} C_{f}}}$$

$$ = R_f \cdot \frac{A_0}{\left( 1 + s R_f C_f \right) \left( 1 + A_0 + \frac{s}{s_p} \right) + \left( 1 + \frac{s}{s_p} \right) s R_f C_{in}}$$

$$ = R_f \cdot \frac{A_0}{ \left( 1 + A_0 \right) + s \left( \frac{1}{s_p} + A_0 R_f C_f + R_f C_{in} \right) + s^2 \frac{R_f \left( C_f + C_{in} \right)}{s_p}}$$

$$ \approx R_f \cdot \frac{ \frac{s_p A_0}{R_f \left( C_f + C_{in} \right)}}{ \frac{s_p A_0}{R_f \left( C_f + C_{in} \right)} + s \frac{1 + s_p R_f \left( A_0 C_f + C_{in} \right)}{R_f \left(C_f + C_{in} \right)} + s^2 } $$

From the general form of a second-order filter,

$$ \omega _0 ^2 \approx \frac{s_p A_0}{R_f \left( C_f + C_{in} \right)} = \frac{\omega _u}{R_f \left( C_f + C_{in} \right)}$$

$$ \omega _0  = \sqrt{\frac{\omega _u}{R_f \left( C_f + C_{in} \right)}}$$

**Part B**

Continuing from the amplifier transfer function derived in part A,

$$ 2 \zeta \omega_0 = \frac{1 + s_p R_f \left( A_0 C_f + C_{in} \right)}{R_f \left(C_f + C_{in} \right)}$$

$$ \zeta = \frac{1 + \omega _u R_f \left(C_f + \frac{C_{in}}{A_0} \right)}{2 R_f \left(C_f + C_{in} \right)} \sqrt{\frac{R_f \left( C_f + C_{in} \right)}{\omega _u}} $$

Assuming $\frac{C_{in}}{A_0} \ll C_f \ll C_{in}$,

$$ \zeta \approx \frac{1 + \omega _u R_f C_f}{2 \sqrt{\omega _u R_f C_{in}}} $$

Assuming $\omega _u R_f C_f \gg 1$,

$$ \zeta \approx \frac{\omega _u R_f C_f}{2 \sqrt{\omega _u R_f C_{in}}} $$

For a Butterworth filter, $\zeta = \frac{1}{\sqrt{2}}$, so

$$ \frac{1}{\sqrt{2}} = \frac{\omega _u R_f C_f}{2 \sqrt{\omega _u R_f C_{in}}} $$

$$ \sqrt{2 \omega _u R_f C_{in}} = \omega _u R_f C_f $$

$$ \frac{1}{R_f C_f} \approx \sqrt{\frac{\omega _u}{2 R_f C_{in}}} $$

**Part C**

Filter specifications: Butterworth filter with resonant frequency $\omega _0 = 2 \pi \cdot 2~MHz$, with $R_f = 10k \Omega$ and $C_{in} = 100~pF \gg C_{f}$.

From part A:

$$ \omega _0  = \sqrt{\frac{\omega _u}{R_f \left( C_f + C_{in} \right)}} \approx \sqrt{\frac{\omega _u}{R_f C_{in}}}$$

From part B:

$$ \frac{1}{R_f C_f} \approx \sqrt{\frac{\omega _u}{2 R_f C_{in}}} $$

$$ C_f \approx \sqrt{\frac{2 C_{in}}{R_f \omega _u}}$$

Specifications are approximately satisfied with $ \omega _u = 2 \pi \cdot 25~MHz$ and $C_f \approx 11~pF$.

**Part D**

Noise sources in the transimpedance amplifier include the op amp and the feedback resistor $R_f$. For a current noise input, the current through $C_{in}$ is approximately zero due to virtual ground at the op amp inverting input, so the entire noise current flows through $R_f$:

$$ v_{out,n} = - R_f i_n $$

The voltage noise of the op amp can be treated as the input to a non-inverting amplifier circuit. Ignoring the bandwidth limitations of the amplifier ($C_f = 0$, $A(s) = A_0$), the voltage noise at the amplifier output due to op amp noise is:

$$ \frac{v_{out,n}}{e_{na}} = 1 + s R_f C_{in} $$

$$ v_{out,n} = \left(1 + s R_f C_{in} \right) e_{na} $$

The voltage noise at the output due to $R_f$ is:

$$ v_{out,n} = e_n = \sqrt{4 k_B R_f T} \approx 12.87 nV/\sqrt{Hz}$$


So, the input-referenced current noise density is

$$ i_n^2 = \left|\frac{1 + s R_f C_in}{R_f} e_{na}\right|^2 + \left| \frac{\sqrt{4 k_B R_f T}}{R_f}\right|^2 = \frac{1 + (\omega R_f C_{in})^2}{R_f^2} e_{na}^2 + \frac{4 k_B T}{R_f}$$

Accounting for the bandwidth of the Butterworth filter designed in part C, the current noise density is:

$$ i_n^2(\omega) \approx \frac{\omega _0^4}{\omega _0^4 + \omega ^4} \cdot i_n^2(0) $$

$$ = \frac{\omega _0^4}{\omega _0^4 + \omega ^4} \left( \frac{1 +  (\omega R_f C_{in})^2}{R_f^2} e_{na}^2 + \frac{4 k_B T}{R_f} \right) $$


```python
import numpy as np
import matplotlib.pyplot as plt

R_f = 10**4
C_in = 10**-10
omg_0 = 2*np.pi*2*10**6    # Filter resonant/cutoff frequency
e_na = 5*10**-9            # Op amp voltage noise
k_b = 1.38*10**-23
T = 300

omega = 2*np.pi*np.logspace(3,8,501)   # Angular frequency, 1 kHz to 100 MHz

i_n = np.sqrt(((omg_0**4)/(omg_0**4 + omega**4))*((((1+omega**2 * R_f**2 *C_in**2)/(R_f**2))*(e_na**2)) + (4*k_b*T)/R_f))

plt.loglog(omega/(2*np.pi),i_n)
plt.title("Noise Current Density")
plt.ylabel("$i_{n}~(A/\sqrt{Hz})$")
plt.xlabel("Frequency (Hz)")

print("Maximum Noise Current Density")
print(np.max(i_n))
print(" ")
print("Frequency of Noise Maximum")
print(omega[np.argmax(i_n)]/(2*np.pi))
print(" ")
print("Noise current density at 1 kHz") # For part e
print(i_n[0])
```

The filter reaches a maximum noise current density of $i_n \sim 4.55 pA/\sqrt(Hz)$ at approximately the filter bandwith of $2 MHz$. The noise analysis plot shows $v_out,n \approx R_f \cdot i_n$.

**Part E**

Shot noise from a $10~\mu A$ input current:

$$i_{sh}^2 = 2 q i_d \Delta f $$
$$i_{sh} \approx 1.79~pA/\sqrt{Hz}$$

At $f = 1 kHz$, the filter noise current density is:

$$ i_n(2 \pi \cdot 1kHz) = \sqrt{\frac{\omega _0^4}{\omega _0^4 + (2 \pi \cdot 1kHz)^4} \left( \frac{1 + ((2 \pi \cdot 1kHz) R_f C_{in})^2}{R_f^2} e_n^2 + \frac{4 k_B T}{R_f} \right)} \approx 1.38~pA/\sqrt{Hz}$$

Noise factor at $f = 1 kHz$:

$$ F = \frac{|i_{sh}|^2 + |i_n(2 \pi \cdot 1kHz)|^2}{|i_{sh}|^2} \approx \frac{|1.79~pA/\sqrt{Hz}|^2 + |1.38~pA/\sqrt{Hz}|^2}{|1.79~pA/\sqrt{Hz}|^2} \approx 1.594$$
