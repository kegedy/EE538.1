**Part A**

Ignoring $r_ \mu$, $C_\mu$, and $C_\pi$, and assuming:

$$ \beta \gg 1 $$
$$ \frac{R_{E2}}{\beta} \ll R_C \ll \beta R_{E2} $$
$$ g_{m2} R_{E2} \gg 1 $$
$$ r_b \approx 0 $$
$$ r_o \gg R_{C}, R_{E2} $$

The amplifier can be split into three parts: (1) the biasing circuit, (2) the common emitter stage, and (3) the common collector (emitter follower) stage. The total transfer function of the circuit is

$$ H(s) = H_{bias}(s) H_{ce}(s) H_{cc}(s) $$

Starting from the output, the common collector stage has gain and input impedance:

$$A_v = H_{cc}(s) \approx \frac{g_{m2} R_{E2}}{1 + g_{m2} R_{E2}} \approx 1 $$
$$r_{in2} = r_{\pi 2} + \left( \beta + 1 \right) R_{E2} \approx \beta R_{E2} $$

The common emitter stage with emitter degeneration has gain and input impedence:

$$A_v = H_{ce}(s) \approx -\frac{\beta R_{L1}}{Z_{in1}} $$
$$ Z_{in1} = r_{\pi 1} + \left(\beta + 1 \right) Z_{E} \approx r_{\pi 1} + \beta Z_{E} $$
$$ ~ $$
$$ R_{L1} \approx \left( R_C \parallel r_{in2} \right) \approx R_C $$
$$ Z_{E} = \left(R_E \parallel \frac{1}{s C_{AC2}} \right) = \frac{R_E}{1 + s C_{AC2} R_E} $$

Defining $\tau_2 \equiv R_E C_{AC2}$,

$$ Z_{E} = \frac{R_E}{1 + s \tau_2} $$

$$ H_{ce}(s) \approx \frac{\beta R_C}{r_{\pi 1} + \frac{\beta R_E}{1 + s \tau_2}} = \frac{\beta R_C \left( 1 + s \tau_2 \right)}{r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E} $$

By KCL at the base of Q1, the transfer function of the bias circuit is:

$$ s C_{in} v_{in} = s C_{AC} v_{b} + \frac{v_b}{R_{B1}} + \frac{v_b}{R_{B2}} + \frac{v_b}{Z_{in1}} = v_{b} \left( s C_{AC} + \frac{1}{R_B \parallel Z_{in1}} \right)$$

$$ H_{bias}(s) = \frac{v_b}{v_{in}} = \frac{ s C_{AC} \left( R_B \parallel Z_{in1} \right)} { 1+ s C_{AC} \left( R_B \parallel Z_{in1} \right)} $$

Defining $\tau_1 \equiv R_B C_{AC}$,

$$ H_{bias}(s) = \frac{s \tau_1 \left[ \frac{r_{\pi 1} + \frac{\beta R_E}{1 + s \tau_2}}{R_B + r_{\pi 1} + \frac{\beta R_E}{1 + s \tau_2}} \right]}{1 + s \tau_1 \left[ \frac{r_{\pi 1} + \frac{\beta R_E}{1 + s \tau_2}}{R_B + r_{\pi 1} + \frac{\beta R_E}{1 + s \tau_2}} \right]} $$

$$ = \frac{s \tau_1 \left[ \frac{r_{\pi 1} \left(1 + s \tau_2 \right) + \beta R_E}{\left( R_B + r_{\pi 1} \right) \left(1 + s \tau_2 \right) + \beta R_E} \right]}{1 + s \tau_1 \left[ \frac{r_{\pi 1} \left(1 + s \tau_2 \right) + \beta R_E}{\left( R_B + r_{\pi 1} \right) \left(1 + s \tau_2 \right) + \beta R_E} \right]} $$

$$ = \frac{s \tau_1 \left[ r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E \right] } {\left( R_B + r_{\pi 1} \right) \left(1 + s \tau_2 \right) + \beta R_E  + s \tau_1 \left[ r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E \right] } $$

The total transfer function of the circuit is

$$ H(s) = H_{bias}(s) H_{ce}(s) H_{cc}(s) $$.

$$ H(s) = \frac{s \tau_1 \left[ r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E \right] } {\left( R_B + r_{\pi 1} \right) \left(1 + s \tau_2 \right) + \beta R_E  + s \tau_1 \left[ r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E \right] } \cdot \frac{\beta R_C \left( 1 + s \tau_2 \right)}{r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E} \cdot \frac{g_{m2} R_{E2}}{1 + g_{m2} R_{E2}} $$






**Part B**

For the midband circuit, the effects of emitter degeneration and input filtering are negligible, so the voltage gain is close to constant at

$$ |A_v| = g_{m1} R_{L1} \cdot \frac{g_{m2} R_{E2}}{1 + g_{m2} R_{E2}} \approx g_{m1} R_C $$

Since the emitter follower has approximately unity gain, any voltage signal at the output of either stage can be scaled by $\frac{1}{A_v}$ to find an equivalent voltage signal at the input. Ignoring 1/f noise:

The base resistances of each BJT and the two resistors remaining in the midband circuit act as thermal voltage noise sources. The noise signal from $r_{b1}$ is applied to the input of the amplifier circuit; $R_C$ and $r_{b2}$ to the output of the common emitter stage; and $R_{E2}$ to the output of the circuit. So, the input-referenced voltage noise from all of these sources is:

$$e_{n,th}^2 = e_{n,b1}^2 + \frac{e_{n,b2}^2 + e_{n,RC}^2 + e_{n,RE2}^2}{A_v^2} = 4 k_B T \left(r_{b1} + \frac{r_{b2} + R_C + R_{E2}}{A_v^2} \right) $$

Each BJT also contributes base-emitter and collector-emitter shot noise current. $i_{n,b1}$ is applied at the input of the amplifier, so it contributes to the input-referenced noise current of the overall circuit. The other current noise sources cause noise voltage at the output, so they contribute to the input-referenced voltage noise. $i_{n,c1}$ and $i_{n,b2}$ are in parallel with $R_C$ and $r_{\pi 2}$, respectively, so they can be converted to voltage signals affecting the output of the common emitter stage by source transformation. $i_{n,c2}$ acts like an output current, so it causes a voltage noise signal at the output based on the circuit output resistance $r_{out}$. So,

$$i_n^2 = i_{n,b1}^2 = 2 q \frac{I_{C1}}{\beta} $$

$$e_{n,shot}^2 = \frac{i_{n,c1}^2 R_C^2}{A_v^2} + \frac{i_{n,b2}^2 R_{\pi 2}^2}{A_v^2} + \frac{i_{n,c2}^2 R_{out}^2}{A_v^2} = 2 q \left( \frac{I_{C1} R_C^2}{A_v^2} + \frac{I_{C2} r_{\pi 2}^2}{\beta A_v^2} + \frac{I_{C2} r_{out}^2}{A_v^2} \right) $$

where

$$r_{out} \approx R_E \parallel \left(\frac{1}{g_{m2}} + \frac{R_C}{\beta} \right)$$



The total input-referenced noise is:

$$ e_n^2 = e_{n,th}^2 + e_{n,shot}^2 = 4 k_B T \left(r_{b1} + \frac{r_{b2} + R_C + R_{E2}}{A_v^2} \right) + 2 q \left( \frac{I_{C1} R_C^2}{A_v^2} + \frac{I_{C2} r_{\pi 2}^2}{\beta A_v^2} + \frac{I_{C2} r_{out}^2}{A_v^2} \right) $$

$$i_n^2 = 2 q \frac{I_{C1}}{\beta} $$


The bias resistors $R_{B1}$ and $R_{B2}$ would also contribute current noise at the input, but these were omitted since the resistors were not included in the midband circuit diagram.






**Part C**

For a BJT,

$$ g_m = \frac{I_C}{V_{th}} $$

$$ r_{\pi} = \frac{\beta}{g_m} = \frac{\beta V_{th}}{I_C} $$ 

where $V_{th}$ is the thermal voltage

$$ V_{th} = \frac{k_B T}{q} \approx 26~mV @ RT$$.


Assuming $g_{m2} R_{E2} \gg 1$ and $ \frac{1}{g_{m2}} \gg \frac{R_C}{\beta}$, 

$$r_{out} \approx \frac{1}{g_{m2}} = \frac{V_{th}}{I_{C2}} = 50~\Omega $$

$$ I_{C2} = \frac{V_{th}}{r_{out}} = \frac{V_{th}}{50~\Omega} \approx 520~\mu A $$


The midband voltage gain of the amplifier is 

$$ A_v \approx g_{m1} R_C = \frac{I_{C1} R_{C}}{V_{th}} = 40dB = 100 $$
$$ I_{C1} = \frac{A_v V_{th}}{R_C} $$



The source resistance converts the amplifier current noise to a voltage noise at the input:

$$e_{n,i}^2 = i_n^2 \left(R_s \parallel r_{\pi 1} \right)^2 =  2 q \frac{I_{C1}}{\beta} \left(R_s \parallel r_{\pi 1} \right)^2 $$

So, the total voltage noise from the amplifier circuit is:

$$e_{n,amp}^2 = 
4 k_B T \left(r_{b1} + \frac{r_{b2} + R_C + R_{E2}}{A_v^2} \right) 
+ 2 q \left( \frac{I_{C1} R_C^2}{A_v^2} 
+ \frac{I_{C1} \left(R_s \parallel r_{\pi 1} \right)^2}{\beta}
+ \frac{I_{C2} r_{\pi 2}^2}{\beta A_v^2} 
+ \frac{I_{C2} r_{out}^2}{A_v^2}   \right) $$


Converting $r_{\pi} \rightarrow \frac{\beta V_{th}}{I_C}$:

$$e_{n,amp}^2 = 4 q V_{th} \left(r_{b1} + \frac{r_{b2} + R_C + R_{E2}}{A_v^2} \right)
+ 2 q \left( \frac{I_{C1} R_C^2}{A_v^2}
+ \frac{I_{C1}}{\beta} \left( \frac{\beta R_s V_{th}}{I_{C1} R_s + \beta V_{th}} \right) ^2
+ \frac{\beta V_{th}^2}{I_{C2} A_v^2}
+ \frac{I_{C2} r_{out}^2}{A_v^2}
 \right) $$


Converting $R_C \rightarrow \frac{A_v V_{th}}{I_{C1}}$ and $I_{C2} \rightarrow \frac{V_{th}}{r_{out}}$,

$$e_{n,amp}^2 = 4 q V_{th} \left(r_{b1} + \frac{r_{b2} + R_{E2}}{A_v^2} + \frac{V_{th}}{A_v I_{C1}} \right)
+ 2 q \left( \frac{ V_{th}^2}{I_{C1}}
+ \frac{I_{C1}}{\beta} \left( \frac{\beta R_s V_{th}}{I_{C1} R_s + \beta V_{th}} \right) ^2
+ \frac{\left(1 + \beta \right) V_{th} r_{out}}{A_v^2}
 \right) $$



The noise from the source resistance is:

$$ e_{n,s}^2 = 4 k_B T R_s = 4 q V_{th} R_s $$

The noise figure of the circuit is:

$$ NF \equiv 10 log \left( \frac{e_{n,s}^2 + e_{n,amp}^2}{e_{n,s}^2} \right) $$

```python
import numpy as np
import matplotlib.pyplot as plt

R_s = 100
r_b = 2
r_out = 50
beta = 250
A_v = 100

R_E2 = 2000             # Arbitrary value satisfying R_E2 >> r_out
V_th = 25.875e-3        # Thermal voltage @ T=300K, V_th = kT/q
q = 1.6e-19             # Fundamental charge

I_C1 = np.logspace(-4,1,501)

en2_s = 4*q*V_th*R_s   # Noise from source resistance

en2_rc = (4*q*V_th**2)/(A_v * I_C1)                   # Noise from R_C (resistor value depends on I_C1)
en2_r = 4*q*V_th*(r_b + (r_b + R_E2)/(A_v**2))        # Noise from other resistors
en2_thermal = en2_rc + en2_r                          # Noise from thermal sources in amplifier


en2_ic1 = ((2*q*V_th**2)/(I_C1))
en2_ib1 = ((2*q*I_C1)/(beta))*(((beta*R_s*V_th)/(I_C1*R_s + beta*V_th))**2)

en2_q1 = en2_ic1 + en2_ib1  # Shot noise from Q1

en2_q2 = (1+beta)*q*V_th*r_out/(A_v**2)  # Shot noise from Q2

en2_amp  = en2_thermal + en2_q1 + en2_q2

NF = 10*np.log10((en2_s + en2_amp)/(en2_s))

#plt.loglog(I_C1, np.ones(np.size(I_C1))*en2_s)
#plt.loglog(I_C1, np.ones(np.size(I_C1))*en2_r)
#plt.loglog(I_C1, np.ones(np.size(I_C1))*en2_q2)
#plt.loglog(I_C1, en2_rc)
#plt.loglog(I_C1, en2_ic1)
#plt.loglog(I_C1, en2_ib1)

plt.loglog(I_C1,NF)
plt.xlabel("Bias Current $I_{C1}$ (A)")
plt.ylabel("Noise Figure")
plt.ylim([0.2,10])

print("Minimum Noise Figure")
print(np.min(NF))
print("")
print("Minimum Noise Bias Current")
print(I_C1[np.argmin(NF)])
print("")
print("Minimum Noise R_C")
print(A_v*V_th/I_C1[np.argmin(NF)])
```

A bias current of $I_{C1} \approx 4.8~mA$ for the common-emitter stage gives a minimum noise figure of $NF \approx 0.35$. With $R_C \approx 540~\Omega$, this bias current gives a midband voltage gain of 40 dB. Technically, the noise figure continues to decrease for higher bias currents, but BJT's that can support collector currents in excess of 1A are not typically used outside of power electronics.






**Part D**

From part C, assuming the same given/target parameters: $R_C = 540~\Omega$, $I_{C1} = 480~\mu A$, $I_{C2} = 520~\mu A$. Assuming $V_{BE} \approx 0.7~V$ and $I_C \approx I_E$ for both BJT's,

$$I_{C2} \approx \frac{V_{B2}-V_{BE}}{R_{E2}} = \frac{V_{CC}-R_C I_{C1}-V_{BE}}{R_{E2}}$$

$$520~\mu A = \frac{9~V - \left( 540~\Omega \cdot 4.8~mA \right) - 0.7~V}{R_{E2}}$$

$$R_{E2} = \frac{5.71~V}{520~\mu A} = 11~k \Omega$$

Note that this value of $R_{E2}$ increases the noise from the amplifier slightly compared to the calculations in part C.

$$I_{C1} \approx \frac{V_{B1} - V_{BE}}{R_E}$$

Arbitrarily assigning $V_{B1} = 3.7~V$, the emitter degeneration resistance $R_E$ is set to provide the specified bias current:

$$R_E = \frac{V_{B1} - V_{BE}}{I_{C1}} = \frac{3~V}{4.8~mA} = 625 \Omega$$

For the DC circuit, assuming $\frac{V_{CC}}{R_{B1}+R_{B2}} \gg I_{B1} = \frac{I_{C1}}{\beta}$, the bias voltage at the base of Q1 is determined by a voltage splitter:

$$V_{B1} = V_{CC} \frac{R_{B2}}{R_{B1} + R_{B2}}$$

$$ R_{B1} = \frac{V_{CC} - V_{B1}}{V_{B1}} R_{B2} = \frac{9~V - 3.7~V}{3.7~V} R_{B2} \approx 1.43 R_{B2}$$

Arbitrarily setting $R_{B1} = 10~k \Omega$, this condition is satisfied with $R_{B2} \approx 7~k \Omega$. The DC current through the bias circuit is $\frac{9~V}{17~k \Omega} \approx 530~\mu A \gg \frac{I_{C1}}{\beta}$. The parallel combination of the bias resistors is $R_B = R_{B1} \parallel R_{B2} \approx 4.1~k \Omega$.

Both the input coupling capacitor $C_{AC}$ and emitter degeneration bypass capacitor $C_{AC2}$ act as high-pass filters, so both values are set to satisfy the high-pass 3dB frequency specification. Temporarily ignoring $R_s$ and using the transfer function $H(s)$ and time constants $\tau_1 = R_B C_{AC}$ and $\tau_2 = R_E C_{AC2}$ from part A:

$$ H(s) = \frac{s \tau_1 \left[ r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E \right] } {\left( R_B + r_{\pi 1} \right) \left(1 + s \tau_2 \right) + \beta R_E  + s \tau_1 \left[ r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E \right] } \cdot \frac{\beta R_C \left( 1 + s \tau_2 \right)}{r_{\pi 1} \left( 1 + s \tau_2 \right) + \beta R_E} \cdot \frac{g_{m2} R_{E2}}{1 + g_{m2} R_{E2}} $$

```python
V_th = 25.875e-3
beta = 250
I_C1 = 4.8e-3

r_pi = beta*V_th/I_C1
R_E = 625
R_C = 540
R_B = (10e3 * 7e3)/(10e3 + 7e3)

# For simplicity, assuming tau_1 = A_v*tau_2 = tau, to match bandwidths of bias circuit and common emitter transfer functions.
# Frequency is defined relative to time constant tau_1.
omega = np.logspace(-1,4,501)
s = 1j*omega
k = 100

H_bias = s*(r_pi*(1+k*s)+beta*R_E)/((R_B+r_pi)*(1+k*s) + beta*R_E + s*(r_pi*(1+k*s) + beta*R_E))

H_ce = beta*R_C*(1+k*s)/(r_pi*(1+k*s) + beta*R_E)

# Assuming unity gain for emitter follower...

H = H_bias*H_ce

H2 = np.abs(H)**2

plt.loglog(omega,np.abs(H_ce))
plt.ylabel("Voltage Gain $A_v$")
plt.xlabel("Normalized Frequency")

plt.figure()
plt.loglog(omega,np.abs(H)/np.max(np.abs(H)))
plt.ylabel("Relative Gain")
plt.xlabel("Normalized Frequency")
plt.loglog(omega,np.abs(H_ce)/np.max(np.abs(H_ce)))
plt.loglog(omega,np.abs(H_bias)/np.max(np.abs(H_bias)))

print("Maximum Gain")
H_max = np.sqrt(np.max(H2))
print(H_max)
print("")
print("Half Power Relative Frequency")
w_3dB = omega[np.argmin(np.abs(H2-(0.5*np.max(H2))))]
print(w_3dB)
```

The amplifier circuit reaches the -3dB high-pass point when $\omega = 2 \pi f \approx \frac{5}{\tau}$. For a corner frequency $f_c = 100~Hz$,

$$\tau_1 \approx \frac{5}{2 \pi \cdot 100~Hz} \approx 8~ms$$

$$\tau_2 \approx \frac{\tau_1}{A_v} \approx 80~\mu s $$

So,

```python
f_c = 100
tau = w_3dB/(2*np.pi*f_c)

C_ac = tau/R_B
C_ac2 = tau/(k*R_E)

print("C_AC")
print(C_ac)
print("")
print("C_AC2")
print(C_ac2)
```

To satisfy the corner frequency specification, $C_{AC} > 2~\mu F$ and $C_{AC2} > 1.27 mF$. The circuit was simulated with $C_{AC} = 2~\mu F$ and $C_{AC2} = 1.3 mF$. Combined with the $100~\Omega$ source resistance and the previously determined component values, the simulated gain of the amplifier was $A_v \approx 38.6~dB \approx 85$, with a 3 dB high-pass corner frequency of $f_c \approx 73.5~Hz$ and a noise figure of $NF \approx 0.47$ at 10 kHz.
